```python
import tensorflow as tf
import numpy as np
import pandas as pd


tf.random.set_seed(42)
X =tf.range(1,1000)
y=X*2+1
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])

model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))

model.fit(X,y,epochs=300)
```

```python
model.predict([-111])
```

```python
import numpy as np
import tensorflow as tf
import keras
from keras import layers
```

```python
inputs = tf.keras.Input(shape=(1,))
outputs = tf.keras.layers.Dense(1)(inputs)
model = keras.Model(inputs=inputs, outputs=outputs)
```

```python
model.summary()
```

```python
model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))

model.fit(X,y,epochs=300)
```

```python
from keras import backend as K
K.clear_session()
class MyModel(tf.keras.Model):
    def __init__(self):
        super(MyModel, self).__init__()
        self.dense = tf.keras.layers.Dense(1, activation="linear")

    def call(self, inputs):
        input_expanded = tf.expand_dims(inputs, axis = 1)
        return self.dense(input_expanded)

model=MyModel()
```

```python
model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))

model.fit(X,y,epochs=300)
```

```python
callback = keras.callbacks.EarlyStopping(monitor='loss', patience=3)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=[1])
])

model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))

model.fit(X,y,epochs=300)
```

```python
early_stopping_callback = tf.keras.callbacks.EarlyStopping(
    monitor='loss',
    patience=10,
    restore_best_weights=True
)
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint("my_checkpoints",
                                                   save_weights_only=True)
model.fit(X,y,epochs=3000,callbacks=[early_stopping_callback, checkpoint_cb])
```
