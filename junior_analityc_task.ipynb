QUIZ 2

```python
average_car_profit_by_supplier = df.groupby('Supplier')['Car profit per reservation'].mean().sort_values(ascending=False)
average_car_profit_by_supplier.head()
```

```python
median_car_profit_by_supplier = df.groupby('Supplier')['Car profit per reservation'].median().sort_values(ascending=False)
median_car_profit_by_supplier.head()
```

```python
top_avg_supplier = average_car_profit_by_supplier.idxmax()
top_avg_profit = average_car_profit_by_supplier.max()

top_median_supplier = median_car_profit_by_supplier.idxmax()
top_median_profit = median_car_profit_by_supplier.max()

top_avg_supplier, top_avg_profit, top_median_supplier, top_median_profit
```

```python
top_suppliers = [top_avg_supplier, top_median_supplier]

plt.figure(figsize=(12, 8))
sns.boxplot(data=df[df['Supplier'].isin(top_suppliers)], x='Supplier', y='Car profit per reservation')
plt.title('Car Profit per Reservation Distribution for Top Suppliers')
plt.xlabel('Supplier')
plt.ylabel('Car Profit per Reservation')
plt.show()
```

```python
average_car_profit_by_country = df.groupby('Country of residence')['Car profit per reservation'].mean().sort_values(ascending=False)
print(f'1 average \n{average_car_profit_by_country.head()}')
```

```python
median_car_profit_by_country = df.groupby('Country of residence')['Car profit per reservation'].median().sort_values(ascending=False)
print(f" 2 median \n{median_car_profit_by_country.head()}")
```

```python
top_avg_country = average_car_profit_by_country.idxmax()
top_avg_country_profit = average_car_profit_by_country.max()

top_median_country = median_car_profit_by_country.idxmax()
top_median_country_profit = median_car_profit_by_country.max()

print(f"Top Country by Average Car Profit: {top_avg_country} with {top_avg_country_profit}")
print(f"Top Country by Median Car Profit: {top_median_country} with {top_median_country_profit}")
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[df['Country of residence'].isin([top_avg_country, top_median_country])],
            x='Country of residence',
            y='Car profit per reservation')
plt.title('Car Profit per Reservation Distribution for Top Countries')
plt.xlabel('Country of Residence')
plt.ylabel('Car Profit per Reservation')
plt.show()
```

```python
df['Car profit ratio'] = df['Car profit per reservation'] / df['Net price per reservation']
df.head()
```

```python
car_profit_ratio_summary = df['Car profit ratio'].describe()
print(car_profit_ratio_summary)

plt.figure(figsize=(10, 6))
sns.histplot(df['Car profit ratio'], kde=True, bins=30)
plt.title('Distribution of Car Profit Ratios')
plt.xlabel('Car Profit Ratio')
plt.ylabel('Frequency')
plt.show()
```

```python
low_profit_reservations = df[df['Car profit ratio'] < df['Car profit ratio'].quantile(0.25)]
high_profit_reservations = df[df['Car profit ratio'] > df['Car profit ratio'].quantile(0.75)]

print(low_profit_reservations.describe())
print("\n\n")
print(high_profit_reservations.describe())
```

```python
low_profit_by_supplier = low_profit_reservations.groupby('Supplier')['Car profit ratio'].mean().sort_values()
low_profit_by_country = low_profit_reservations.groupby('Country of residence')['Car profit ratio'].mean().sort_values()

high_profit_by_supplier = high_profit_reservations.groupby('Supplier')['Car profit ratio'].mean().sort_values()
high_profit_by_country = high_profit_reservations.groupby('Country of residence')['Car profit ratio'].mean().sort_values()

print(low_profit_by_supplier.head())
print(low_profit_by_country.head())
print(high_profit_by_supplier.head())
print(high_profit_by_country.head())
```
